## Title
Grocery Store Checkout Model

---

## Goal
How long does it take to check out at a grocery store.
When should more cashiers be recruited form the back.

---

## Background
One or more cashiers at any given time.
People must form queue to checkout

---

## Assumptions
The only way to leave the queue is by checking out
Length of queues never differ by more than 1 person

---

## Random Variables
Cashier checkout speed. (For now fixed for each cashier)
Number of items per customer. (For now same # of items per person)
Some customers have coupons. (No coupons to start)
Random talkers. (Nobody talks for initial model)

---

## Experiments
Given an acceptable wait time (5 - 10 minutes)(OR line length)
When is the appropriate time to add/remove cashiers

---

## Initial Conditions
Number of cashiers: 1
Speed of cashiers: 2 Minutes per Person OR 1 Second per Item
Number of customers checking out: 10
Rate of new customers joining the queue: 1 per minute
 
 ---

## Model Elements
Customer Joining/Exiting Queue.
Customer checking out
Adding/Removing cashiers

Later could add:
Different checkout types: express, self checkout, regular
Waves of customers by waves (Time of day, Day of week,Holiday, Random spikes)
